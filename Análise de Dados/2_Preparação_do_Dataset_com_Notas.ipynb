```python
import pandas as pd
```

```python
dados_problemas = pd.read_csv("https://raw.githubusercontent.com/MachineTeachingEdu/JAI2025-IA-Educacao/refs/heads/main/An%C3%A1lise%20de%20Dados/dados_final_problemas.csv", sep=',')
dados_problemas.head()
```

```python
dados_habilidades = pd.read_csv("https://raw.githubusercontent.com/MachineTeachingEdu/JAI2025-IA-Educacao/refs/heads/main/An%C3%A1lise%20de%20Dados/dados_final_habilidades.csv", sep=',')
dados_habilidades.head()
```

```python
ids = dados_habilidades["Id"]
notas = dados_habilidades.drop("Id", axis=1)
notas = notas.mean(axis=1)
notas = notas.rename("Notas")
dados_notas = pd.concat([ids, notas], axis=1)
dados_notas.head()
```

```python
dados_final = dados_problemas.merge(right=dados_notas, on='Id')
dados_final.head()
```

```python
dados_final.to_csv('dados_final.csv', index=False)
```
